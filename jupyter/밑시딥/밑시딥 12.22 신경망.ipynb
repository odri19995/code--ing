```python
# 계단함수의 그래프

import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x>0, dtype= int)

x= np.arange(-5.0,5.0,0.1) #-5.0 부터 5.0  까지 0.1 간격으로 생성
y= step_function(x)
plt.plot(x,y) #x, y 로 그래프 만들기
plt.ylim(-0.1,1.1) # y축의 범위 지정
plt.show()# r그래프 보여주기
```

### 시그모이드 함수 구현하기

```python
def sigmoid(x):
    return 1/ (1 +np.exp(-x))
```

```python
x = np.array([-1.0,1.0,2.0])
sigmoid(x)
```

```python
x =np.arange(-5.0,5.0,0.1)
y =sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

### 비선형 함수 

두 함수 모두 비선형 함수입니다.  선형함수를 활성화 함수로 쓸 수 없는 이유는 선형 함수를 이용하면 신경망의 층을 깊게 하는 의미가 없기 때문입니다. 

###  RelU 함수


```python
def relu(x):
    return np.maximum(0,x)
```

```python
x =np.arange(-1.0,1.0,0.1)
y =relu(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```
